```python
sub_cols = ['pos', 'category', 'w1200', 'w300', 'w120', 'w60', 'w30', 'w15', 'w5',
            'avg_wkg', 'wkg_ftp', 'wftp', 'wkg1200', 'wkg300', 'wkg120', 'wkg60', 'wkg30', 'wkg15','wkg5']

#df = df[sub_cols]
df.head()
```

```python
def clean_w(val):
    try:
        res = int(val[0])
    except:
        res = float(val[0])
    return res
```

```python
for col in sub_cols:
    if col not in  ['pos', 'category']:
        try:
            df.loc[:, col] = df[col].apply(lambda val : clean_w(val))
        except:
            pass
        
df = df.sort_values(by='pos')
df.head()
```

```python
watts_cols = ['w1200', 'w300', 'w120', 'w60', 'w30', 'w15', 'w5']
watts_keys = ['20min', '5min', '2min', '1min', '30sec', '15sec', '5sec']
watts_names = {}
for idx in range(len(watts_cols)):
    watts_names[watts_cols[idx]] = watts_keys[idx]
    

wkg_cols = ['avg_wkg', 'wkg_ftp', 'wkg1200', 'wkg300', 'wkg120', 'wkg60', 'wkg30', 'wkg15', 'wkg5']
wkg_keys = ['Average', 'FTP', '20min', '5min', '2min', '1min', '30sec', '15sec', '5sec']
wkg_names = {}
for idx in range(len(wkg_cols)):
    wkg_names[wkg_cols[idx]] = wkg_keys[idx]
    
```

```python
import dash
import dash_html_components as html
import dash_core_components as dcc
from dash.dependencies import Input, Output
import pandas as pd
import plotly.graph_objs as go
import plotly.express as px

tabs_styles = {
    'height': '44px',
    'align-items': 'center'
}
tab_style = {
    'borderBottom': '1px solid #d6d6d6',
    'padding': '6px',
    'fontWeight': 'bold',
    'border-radius': '15px',
    'background-color': '#F2F2F2',
    'box-shadow': '4px 4px 4px 4px lightgrey',
 
}
 
tab_selected_style = {
    'borderTop': '1px solid #d6d6d6',
    'borderBottom': '1px solid #d6d6d6',
    'backgroundColor': '#119DFF',
    'color': 'white',
    'padding': '6px',
    'border-radius': '15px',
}

app = dash.Dash()


# Create a Dash layout
app.layout = html.Div([
    html.Div(
        html.H1('Zwiftpower Overview', style={"color": 'blue','marginLeft': 100,"textAlign": "left"})
    ),
    dcc.Tabs(id='tabs', value='Tab1', children=[
        dcc.Tab(label='Watts',style = tab_style, selected_style = tab_selected_style,
                id='id_tab1', value='val_tab1', 
                
                children =[            
                    html.H3("Select Category",style={"color": 'green','marginLeft': 100}),
                    html.Div(
                        [
                            dcc.Dropdown(
                                id="id_watts",
                                options=[{ 'label': i,'value': i } for i in cat_cols],value='A'),
                        ],style={'width': '15%', 'display': 'inline-block','marginLeft': 100}),
                    
                html.H2("Watts Values for each Postion",  style={"color":"red","textAlign": "center","marginTop":60}),
                dcc.Graph(id="app_1"),
                
                html.H2("\nOverall Watt Insights",  style={"color": 'red','marginTop': 50,"textAlign": "center"}),
                dcc.Graph(
                id='cloudwatch',
                figure={
                    'data': [
                        {'x': df['pos'], 'y': df['w1200'], 'type': 'line', 'name': '20min'},
                        {'x': df['pos'], 'y': df['w300'], 'type': 'line', 'name': '5min'},
                        {'x': df['pos'], 'y': df['w120'], 'type': 'line', 'name': '2min'},
                        {'x': df['pos'], 'y': df['w30'], 'type': 'line', 'name': '30sec'},
                        {'x': df['pos'], 'y': df['w15'], 'type': 'line', 'name': '15min'},
                        {'x': df['pos'], 'y': df['w5'], 'type': 'line', 'name': '5sec'},
                    ],
                    'layout' : {"width": '50%','yaxis_range' : [0,max(df[watts_cols])]}, 'height' : 600,
                }),
                
        ]),
        
        dcc.Tab(label='WKG',style = tab_style, selected_style = tab_selected_style,
                 id='id_tab2', value='val_tab2', 
                 
                children=[
                     html.H3("Select Category",style={"color": 'green','marginLeft': 100}),
                     html.Div(
                        [
                            dcc.Dropdown(
                                id="id_wkg",
                                options=[{'label': i, 'value': i} for i in cat_cols], value='A'),],
                                    style={'width': '15%', 'display': 'inline-block','marginLeft': 100}),
                
                    html.H2("WKG Values for each Postion",  style={"color":"red","textAlign": "center","marginTop":60}),
                dcc.Graph(id="app_2"),
                
                html.H2("\nOverall WKG Insights",  style={"color": 'red','marginTop': 50,"textAlign": "center"}),
                dcc.Graph(
                id='cloudwatch_wkg',
                figure={
                    'data': [
                        #'wkg1200', 'wkg300', 'wkg120', 'wkg60', 'wkg30', 'wkg15', 'wkg5'
                        {'x': df['pos'], 'y': df['wkg1200'], 'type': 'line', 'name': '20min'},
                        {'x': df['pos'], 'y': df['wkg300'], 'type': 'line', 'name': '5min'},
                        {'x': df['pos'], 'y': df['wkg120'], 'type': 'line', 'name': '2min'},
                        {'x': df['pos'], 'y': df['wkg60'], 'type': 'line', 'name': '60sec'},
                        {'x': df['pos'], 'y': df['wkg30'], 'type': 'line', 'name': '30sec'},
                        {'x': df['pos'], 'y': df['wkg15'], 'type': 'line', 'name': '15sec'},
                        {'x': df['pos'], 'y': df['wkg5'], 'type': 'line', 'name': '5sec'},
                    ],
                    'layout' : {"width": '50%','yaxis_range' : [0,max(df[wkg_cols])]}, 'height' : 600,
                })
                    
        ])
        
    ], style = tabs_styles)
])

@app.callback(
    [dash.dependencies.Output('app_1', 'figure')],
    [dash.dependencies.Input('id_watts', 'value')])

def update_graph(id_watts):
    df_filter = df[df['category']==id_watts]
    df_filter = df_filter.sort_values(by='pos')
        
    fig_1 = px.line(df_filter, x="pos", y=watts_cols, markers='o', height=600,
                    labels=dict(x="Position", y="Watts Values"))
    fig_1['data'][0]['showlegend']=True
    
    fig_1.update_layout(
           xaxis_title = "Postion",
           yaxis_title = "Category : "+id_watts,
           font = dict(family = "Courier New", size = 14, color = "blue"))
    
    fig_1.update_xaxes(ticks="outside", tickwidth=2, tickcolor='crimson', ticklen=10)
    fig_1.update_yaxes(ticks="outside", tickwidth=2, tickcolor='crimson', ticklen=10, col=1, range = [0,df_filter[watts_cols]])
    fig_1.update_xaxes(tickfont=dict(family='Rockwell', color='crimson', size=16))
    return [fig_1]

@app.callback(
    [dash.dependencies.Output('app_2', 'figure')],
    [dash.dependencies.Input('id_wkg', 'value')])

def update_graph(id_wkg):   
    df_filter = df[df['category']==id_wkg]
    df_filter = df_filter.sort_values(by='pos')
    
    fig_2 = px.line(df_filter, x="pos", y=wkg_cols, markers='o', height=600,
                    labels=dict(x="Position", y="WKG Values"))
    fig_2['data'][0]['showlegend']=True
    #fig_2['data'][0]['name']=id_wkg+' : value'
    fig_2.update_layout(
           xaxis_title = "Postion",
           yaxis_title = "Category : "+id_wkg,
           font = dict(family = "Courier New", size = 14, color = "blue"))
    
    fig_2.update_xaxes(ticks="outside", tickwidth=2, tickcolor='crimson', ticklen=10)
    fig_2.update_yaxes(ticks="outside", tickwidth=2, tickcolor='crimson', ticklen=10, col=1,range = [0,df_filter[wkg_cols]])
    fig_2.update_xaxes(tickfont=dict(family='Rockwell', color='crimson', size=16))
    return [fig_2]



if __name__ == '__main__':
    app.run_server(debug=False)
```
