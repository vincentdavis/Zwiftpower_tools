```python
from zpdatareader import FetchJson
from ZFileDb import ZFileDb

# Define the database
db = ZFileDb(db_path="database/ZFileDb")
z = FetchJson(db=db)
z.login()
```

```python
# Get the team list
teams = z.fetch_teamlist(refresh=True)
```

```python
# Sort the teams by rider count
def to_int(v):
    try:
        v = int(v)
    except:
        v=0
    finally:
        return v

teams['teamlist'] = sorted(teams['teamlist'], key = lambda x: to_int(x['riders']), reverse=True)

# look  at the 1000th team
z.fetch_team(teams['teamlist'][1000]['team_id'])
```

```python
# Get the top (rider count) 400 teams.
# This is slow and that is good.
import time
for t in teams['teamlist'][000:400]:
    r, c  = z.fetch_team(t['team_id'])
    print(t['team_id'], t['riders'], c)
    if c == 'refresh':
        time.sleep(15)
print('Done')
```
